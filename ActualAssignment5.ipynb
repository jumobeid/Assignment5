```python
def visualize(filename):
    "Plot the subplot"
    data = pd.read_csv(filepath_or_buffer = filename)
    a_list = list(range(0, 40))
    x = np.array(a_list)
    i = list(range(0, 3))
    counter = 1
    for i,name,y in [(1,'Average', data.mean(axis=0)),
                    (2,'Max', data.max(axis=0)),
                    (3,'Min', data.mean(axis=0))]: 
   
        #print(x)
        #print(y)
        if i == 1:
            select_color = "yellow"
        elif i == 2:
            select_color = "Green"
        else: 
            select_color = "red"
        plt.subplot(3,1,i)
        plt.plot(x,y ,color= select_color)
        plt.title("dataset" + filename[16:17])
        plt.xlabel("Feature")
        plt.ylabel(name)
        matplotlib.pyplot.show()
    
```
