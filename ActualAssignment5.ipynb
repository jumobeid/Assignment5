```python
def visualize(filename):
    "Plot the subplot"
    data = pd.read_csv(filepath_or_buffer = filename)
    a_list = list(range(0, 40))
    x = np.array(a_list)
    y = np.array(data.mean(axis=0))
    #print(x)
    #print(y)
    plt.subplot(1, 2, 1)
    plt.plot(x,y)
    matplotlib.pyplot.show()
    
```

```python
def stat(filename):
    "Print the descriptive statistics"
    data = pd.read_csv(filepath_or_buffer = filename)
    print(data.mean(axis=0))
```
